```python
data = pd.read_csv("data/final_dataSet.csv")

data.head()
```

```python
data["State_Name"].replace(
    {'Jammu and Kashmir':0, 'Jharkhand':1, 'Bihar':2, 'Uttarakhand':3, 'Chhattisgarh':4, 'Punjab':5, 'Arunachal Pradesh':6, 'Odisha':7, 'Kerala':8, 'Himachal Pradesh':9, 'Tamil Nadu':10, 'Andaman and Nicobar Islands':11},
    inplace=True
)
```

```python
x = dict.fromkeys(set(data["Crop"]), 0)
print(type(x))
```

```python
index = 0
for i in x:
    x[i] = index
    index += 1
x
```

```python
data["Crop"].replace(x, inplace=True)
```

```python
data.dropna()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

import numpy as np
```

```python
scaler =  MinMaxScaler()
data[["Area", "Production", "Rain_fall"]] = scaler.fit_transform(data[["Area", "Production", "Rain_fall"]])

data = data.dropna()
```

```python
data
```

```python
train_data = data.drop(['Rain_fall'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(train_data, data['Rain_fall'], random_state=0)

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
prediction = lr.predict(X_test)
print(y_test)
print(prediction)
```

```python
import pickle

with open("model_pickel", "wb") as f:
    pickle.dump(lr, f)
```
